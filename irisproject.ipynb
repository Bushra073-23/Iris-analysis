#### Step 6: **Correlation Analysis**

Expacted code:

```python
corr = df.corr() 
sns.heatmap(corr, annot=True, cmap='coolwarm') 
```

***What's causing it?***
You're trying to compute df.corr(), but your DataFrame (df) contains a non-numeric column — specifically, the species column, which contains string values like "setosa", "versicolor", etc.

***Internally:***
.corr() tries to convert all columns to numeric.

When it reaches the species column (which is type object/string), it fails with:

could not convert string to float: 'setosa'

***How to Fix It:***
You need to exclude the non-numeric columns when calculating correlation.

#### Corrected Code:

```python
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

High values like 0.96 between petal_length and petal_width indicate a very strong correlation.

***Objective:***

Measure how strongly features are related to each other (statistically).

Use a heatmap to visualize these relationships easily.

- Create a combined visualization (like sns.FacetGrid) showing multiple plots together

- Create a custom matplotlib subplot layout combining histograms and scatterplots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a FacetGrid for scatter plots
g = sns.FacetGrid(df, col='species', height=4)
g.map_dataframe(sns.scatterplot, x='petal_length', y='petal_width')
g.set_axis_labels("Petal Length", "Petal Width")
g.fig.suptitle("Petal Length vs Width by Species", y=1.05)
plt.show()
```

***Explanation:***

FacetGrid: Creates a layout of plots per category (species).

map_dataframe: Applies a plot function (like scatterplot) to each subset.

This helps compare the same plot across species side-by-side.



```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# Histogram of sepal length
axs[0].hist(df['sepal_length'], bins=20, color='skyblue', edgecolor='black')
axs[0].set_title("Histogram of Sepal Length")
axs[0].set_xlabel("Sepal Length (cm)")
axs[0].set_ylabel("Frequency")

# Scatter plot of petal length vs width
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df, ax=axs[1])
axs[1].set_title("Petal Length vs Width")
axs[1].set_xlabel("Petal Length (cm)")
axs[1].set_ylabel("Petal Width (cm)")

plt.tight_layout()
plt.show()
```

***Explanation:***
    
plt.subplots(): Sets up multiple plots in one figure.

axs[0] and axs[1]: Let you control each subplot.

Combines distribution and relationship analysis in one view.
